```python
import os
import time
import re
import glob
import numpy as np
import pandas as pd
import pickle as pkl
```

glob:  We can pass in a pattern to glob(), including wildcard characters, and it will return a list of all files that match that pattern.

```python
main_dir = "/content/drive/MyDrive/Dataset/Learning_set"
bearing = ['Bearing1_1', 'Bearing1_2', 'Bearing2_1', 'Bearing2_2', 'Bearing3_1', 'Bearing3_2']

DATA_POINTS_PER_FILE = 2560

header_col = ['hour',	'minute',	'second',	'microsecond',	'horiz accel',	'vert accel']
```

```python
for root,dir,lis in os.walk(main_dir, topdown=True):
  print(root," ",dir," ",lis)
  print("total file ",len(lis))
  files_acc = len(glob.glob(os.path.join(root,'acc_*.csv')))
  files_temp = len(glob.glob(os.path.join(root,'temp_*.csv')))
  print("acc files : ", files_acc, " | temp files : ", files_temp,"\n")
```

```python
def files_merge():

  for root,dir,lis in os.walk(main_dir, topdown=True):

    if len(lis) == 0:
      continue
    
    bearing = re.findall('Bearing+[\w.]+',root)[0]
    
    files = sorted(glob.glob(os.path.join(root,'acc_*.csv')))

    merge_tocsv(bearing,files)
    print("file merge successful", bearing ," csv file created")
  
    csv_topkz(bearing)
    print(bearing, ".pkz file created")

def merge_tocsv(bearing_name, file_lis):
  
    dir = "/content/drive/MyDrive/FEMO2012 CHALLENGE /Dataset/csv files" 
    bearing_name = bearing_name + ".csv"
    csv_file = os.path.join(dir,bearing_name)

    with open(csv_file, 'w') as wr:
        for file in file_lis:
            with open(file , 'r') as read:
              for line in read:
                wr.write(line)

def csv_topkz(bearing):
  
  dir_csv = "/content/drive/MyDrive/FEMO2012 CHALLENGE /Dataset/csv files"
  dir_pkz = "/content/drive/MyDrive/FEMO2012 CHALLENGE /Dataset/pkz files"

  csv_file = os.path.join( dir_csv,bearing + ".csv")
  pkz_file = os.path.join( dir_pkz,bearing + ".pkz")

  with open(pkz_file, 'wb') as pkz:
    df = pd.read_csv(csv_file, names= header_col)
    pkl.dump(df, pkz)
            

```

```python
files_merge()
```

```python
# checking if the pkz file is same as csv file 
csv_dir = '/content/drive/MyDrive/FEMO2012 CHALLENGE /Dataset/csv files'
pkz_dir = '/content/drive/MyDrive/FEMO2012 CHALLENGE /Dataset/pkz files'
for b in bearing:

  d1 = pd.read_csv(os.path.join(csv_dir,b+'.csv'), names = header_col)

  with open(os.path.join(pkz_dir,b+'.pkz'), 'rb') as file:
    d2 = pkl.load(file)
    print(d2.shape[0] ,":" ,b)
  
  print(d1.equals(d2),"\n")
```

```python
header_col = ['hour', 'minute', 'second', 'microsecond', 'temp']
def files_merge():

  for root,dir,lis in os.walk(main_dir, topdown=True):

    if len(lis) == 0:
      continue
    
    bearing = "temp_" + re.findall('Bearing+[\w.]+',root)[0]
    
    files = sorted(glob.glob(os.path.join(root,'temp_*.csv')))
    if(len(files) == 0):
      continue

    merge_tocsv(bearing,files)
    print("file merge successful", bearing ," csv file created")
  
    csv_topkz(bearing)
    print(bearing, ".pkz file created")

def merge_tocsv(bearing_name, file_lis):
  
    dir = "/content/drive/MyDrive/FEMO2012 CHALLENGE /Dataset/temp_csv files" 
    bearing_name = bearing_name + ".csv"
    csv_file = os.path.join(dir,bearing_name)

    with open(csv_file, 'w') as wr:
        for file in file_lis:
            with open(file , 'r') as read:
              for line in read:
                wr.write(line)

def csv_topkz(bearing):
  
  dir_csv = "/content/drive/MyDrive/FEMO2012 CHALLENGE /Dataset/temp_csv files"
  dir_pkz = "/content/drive/MyDrive/FEMO2012 CHALLENGE /Dataset/temp_pkz files"

  csv_file = os.path.join( dir_csv,bearing + ".csv")
  pkz_file = os.path.join( dir_pkz,bearing + ".pkz")

  with open(pkz_file, 'wb') as pkz:
    df = pd.read_csv(csv_file,  sep = '[,;]', names = header_col, engine='python')
    pkl.dump(df, pkz)
            

```
